Question 1: Are there certain odors that were more common with edible or poisonous mushrooms?

```python
#count the number of poisonous and edible mushrooms under each odor
#odor: almond=a,anise=l,creosote=c,fishy=y,foul=f,musty=m,none=n,pungent=p,spicy=s
odor_name = 'odor'
odor_data = data['odor']
class_name = 'class'
class_data = data['class']
odor_df = pd.DataFrame({class_name: class_data, odor_name: odor_data})

edible_mushrooms = odor_df[odor_df['class'] == 'e']
poisonous_mushrooms = odor_df[odor_df['class'] == 'p']
e_almond = (edible_mushrooms['odor'] == 'a').sum()
p_almond = (poisonous_mushrooms['odor'] == 'a').sum()

e_anise = (edible_mushrooms['odor'] == 'l').sum()
p_anise = (poisonous_mushrooms['odor'] == 'l').sum()

e_creosote = (edible_mushrooms['odor'] == 'c').sum()
p_creosote = (poisonous_mushrooms['odor'] == 'c').sum()

e_fishy = (edible_mushrooms['odor'] == 'y').sum()
p_fishy = (poisonous_mushrooms['odor'] == 'y').sum()

e_foul = (edible_mushrooms['odor'] == 'f').sum()
p_foul = (poisonous_mushrooms['odor'] == 'f').sum()

e_musty = (edible_mushrooms['odor'] == 'm').sum()
p_musty = (poisonous_mushrooms['odor'] == 'm').sum()

e_odorless = (edible_mushrooms['odor'] == 'n').sum()
p_odorless = (poisonous_mushrooms['odor'] == 'n').sum()

e_pungent = (edible_mushrooms['odor'] == 'p').sum()
p_pungent = (poisonous_mushrooms['odor'] == 'p').sum()

e_spicy = (edible_mushrooms['odor'] == 's').sum()
p_spicy = (poisonous_mushrooms['odor'] == 's').sum()


#Odor count dict
odor_count = {
    'Category': ['Odorless', 'Almond', 'Anise', 'Creosote', 'Fishy', 'Foul', 'Musty', 'Pungent', 'Spicy'],
    'Edible': [e_odorless, e_almond, e_anise, e_creosote, e_fishy, e_foul, e_musty, e_pungent, e_spicy],
    'Poisonous': [p_odorless, p_almond, p_anise, p_creosote, p_fishy, p_foul, p_musty, p_pungent, p_spicy]
}

#make a dataframe
odor_count_df = pd.DataFrame(odor_count)


#visualization
num_categories = len(odor_count_df)
bar_width = 0.35
x = np.arange(num_categories)

plt.bar(x - bar_width/2, odor_count_df['Edible'], width=bar_width, label='Edible', color='green')
plt.bar(x + bar_width/2, odor_count_df['Poisonous'], width=bar_width, label='Poisonous', color='purple')
plt.xticks(x, odor_count_df['Category'], fontsize=8)
plt.legend()
```

Question 2: Which type of habitat had the most edible mushrooms?

```python
#count the number of edible and poisonous mushrooms in each habitat
#habitat: grasses=g,leaves=l,meadows=m,paths=p,urban=u,waste=w,woods=d
habitat_name = 'habitat'
habitat_data = data['habitat']
class_name = 'class'
class_data = data['class']
habitat_df = pd.DataFrame({class_name: class_data, habitat_name: habitat_data})

edible_mushrooms = habitat_df[habitat_df['class'] == 'e']
poisonous_mushrooms = habitat_df[habitat_df['class'] == 'p']
e_grasses = (edible_mushrooms['habitat'] == 'g').sum()
p_grasses = (poisonous_mushrooms['habitat'] == 'g').sum()

e_leaves = (edible_mushrooms['habitat'] == 'l').sum()
p_leaves = (poisonous_mushrooms['habitat'] == 'l').sum()

e_meadows = (edible_mushrooms['habitat'] == 'm').sum()
p_meadows = (poisonous_mushrooms['habitat'] == 'm').sum()

e_paths = (edible_mushrooms['habitat'] == 'p').sum()
p_paths = (poisonous_mushrooms['habitat'] == 'p').sum()

e_urban = (edible_mushrooms['habitat'] == 'u').sum()
p_urban = (poisonous_mushrooms['habitat'] == 'u').sum()

e_waste = (edible_mushrooms['habitat'] == 'w').sum()
p_waste = (poisonous_mushrooms['habitat'] == 'w').sum()

e_woods = (edible_mushrooms['habitat'] == 'd').sum()
p_woods = (poisonous_mushrooms['habitat'] == 'd').sum()

#Habitat count dict
habitat_count = {
    'Category': ['Grasses', 'Leaves', 'Meadows', 'Paths', 'Urban', 'Waste', 'Woods'],
    'Edible': [e_grasses, e_leaves, e_meadows, e_paths, e_urban, e_waste, e_woods],
    'Poisonous': [p_grasses, p_leaves, p_meadows, p_paths, p_urban, p_waste, p_woods]
}

#make a dataframe
habitat_count_df = pd.DataFrame(habitat_count)


#visualization
num_categories = len(habitat_count_df)
bar_width = 0.35
x = np.arange(num_categories)

plt.bar(x - bar_width/2, habitat_count_df['Edible'], width=bar_width, label='Edible', color='green')
plt.bar(x + bar_width/2, habitat_count_df['Poisonous'], width=bar_width, label='Poisonous', color='purple')
plt.xticks(x, habitat_count_df['Category'], fontsize=8)
plt.legend()
```
